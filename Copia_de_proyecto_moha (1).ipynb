  **Proyecto Big Data: Anlisis de Morosidad en Prstamos Bancarios**

  **Objetivo del Proyecto**:

El objetivo del proyecto es identificar patrones y factores que influyen en la morosidad de los prstamos bancarios. Esto permitir a las entidades financieras anticipar el riesgo de impago y tomar medidas preventivas para reducir la morosidad.

 **Descripcin del Problema**
La morosidad en los prstamos bancarios es un desafo para el sector financiero. Puede afectar la estabilidad financiera de las entidades, aumentar los costos operativos y reducir la liquidez. Entender los factores que llevan al impago ayudar a minimizar estos riesgos.

 **Datos Utilizados**
Datos histricos de clientes, Datos del prstamo; monto, tipo de prstamo, tasa de inters, Estado de pagos etc.

**Technicas de analisis**

  *Entender la distribucin de variables clave (genero, edad, puntaje de crdito,monto del prstamo, ingresos, estado del prestamo etc.)

  *Prediccin de incumplimiento de prstamos: Usar la columna **estado de
  prestamo** como objetivo.

  *Segmentacin de clientes: Agrupar a los solicitantes segn su edad y sus sus caractersticas financieras.

  *Anlisis de correlacin: Evaluar qu factores tienen ms relacin con el incumplimiento (por ejemplo, la tasa de inters, el puntaje de crdito o tipo de credito).

**PROCEDEMOS A LA LIMPIEZA DEL PROYECTO**

```python
nulos = df.isnull().sum()
nulos
```

```python
# Rellenar columnas numricas con la media
df.fillna({
    'rate_of_interest': df['rate_of_interest'].mean(),
    'Interest_rate_spread': df['Interest_rate_spread'].mean(),
    'Upfront_charges': df['Upfront_charges'].mean(),
    'term': df['term'].mean(),
    'property_value': df['property_value'].mean(),
    'income': df['income'].mean(),
    'LTV': df['LTV'].mean(),
    'dtir1': df['dtir1'].mean()
}, inplace=True)

# Rellenar columnas categricas con la moda
df.fillna({
    'loan_limit': df['loan_limit'].mode()[0],
    'approv_in_adv': df['approv_in_adv'].mode()[0],
    'loan_purpose': df['loan_purpose'].mode()[0],
    'Neg_ammortization': df['Neg_ammortization'].mode()[0],
    'age': df['age'].mode()[0],
    'submission_of_application': df['submission_of_application'].mode()[0]
}, inplace=True)

# Verificar si se han reemplazado los valores nulos
valores_nulos = df.isnull().sum()
df.to_csv("Loan_Default_Limpia.csv", index=False)

valores_nulos

```

**VAMOS A INTENTAR AGRUPAR ALGUNAS COLUMNAS CLAVE CON LA COLUMNA OBJETIVO (ESTADO DE PRESTAMO) PARA ANALIZAR QUE FACTOR INFLUE MAS AL IMPAGO DEL PRESTAMO**

```python
df_original = pd.DataFrame(df)
columnas_seleccionadas = ['Gender','Status']
df_nuevo = df_original[columnas_seleccionadas]
df_nuevo
```

```python
import pandas as pd
from matplotlib import pyplot as plt

# Agrupamos los datos por 'Gender' y 'Status', luego contamos las ocurrencias
df_grouped = df_nuevo.groupby(['Gender', 'Status']).size().unstack()

# Cambiamos los nombres de las columnas (Status) de 0 a 'Pagado' y de 1 a 'No Pagado'
df_grouped.columns = ['Pagado' if x == 0 else 'No Pagado' for x in df_grouped.columns]

# Graficamos
df_grouped.plot(kind='bar', stacked=True)
plt.title('Distribucin del estado de prestamo por gnero')
plt.ylabel('Frencuencia')
plt.xlabel('Gnero')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.legend(title='Estado')
plt.show()
```
